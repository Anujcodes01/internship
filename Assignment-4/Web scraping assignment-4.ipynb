```python
# connecting to the webdriver 
os.environ['PATH']+= r"C:\Users\ASUS\Documents\internship\chormedriver.exe"
driver = webdriver.Chrome()
driver.get('https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos')

```

```python
# create an empty list to store the video names
names = []
# find all table cells with column index 2 (i.e., the video name column) and extract the text from each cell
# then add the text to the names list after removing any leading/trailing white space
for i in driver.find_elements(By.XPATH, "//table[@class='wikitable sortable jquery-tablesorter'][1]/tbody/tr/td[2]"):
    names.append(i.text.strip())

# create an empty list to store the video ranks
Rank = []
# find all table cells with column index 1 (i.e., the video rank column) and extract the text from each cell
# then add the text to the Rank list after removing any leading/trailing white space
for i in driver.find_elements(By.XPATH, "//table[@class='wikitable sortable jquery-tablesorter'][1]/tbody/tr/td[1]"):
    Rank.append(i.text.strip())
    
# create an empty list to store the video artists
Artist = []
# find all table cells with column index 3 (i.e., the video artist column) and extract the text from each cell
# then add the text to the Artist list after removing any leading/trailing white space
for i in driver.find_elements(By.XPATH, "//table[@class='wikitable sortable jquery-tablesorter'][1]/tbody/tr/td[3]"):
    Artist.append(i.text.strip())

# create an empty list to store the video upload dates
Upload_date = []
# find all table cells with column index 5 (i.e., the video upload date column) and extract the text from each cell
# then add the text to the Upload_date list after removing any leading/trailing white space
for i in driver.find_elements(By.XPATH, "//table[@class='wikitable sortable jquery-tablesorter'][1]/tbody/tr/td[5]"):
    Upload_date.append(i.text.strip())

# create an empty list to store the video views
Views = []
# find all table cells with column index 4 (i.e., the video views column) and extract the text from each cell
# then add the text to the Views list after removing any leading/trailing white space
for i in driver.find_elements(By.XPATH, "//table[@class='wikitable sortable jquery-tablesorter'][1]/tbody/tr/td[4]"):
    Views.append(i.text.strip())  
    
# create a Pandas DataFrame with the video names, artists, upload dates, and views
# use the Rank list as the index for the DataFrame
df = pd.DataFrame({'Video Name': names, 'Artist': Artist, 'Upload Date': Upload_date, 'Views': Views}, index=Rank)

# display the DataFrame
df
```

```python
# Importing the necessary libraries
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.common.exceptions import NoSuchElementException
import pandas as pd

# Creating a webdriver instance for Chrome
driver = webdriver.Chrome("chromedriver.exe")

# Navigating to the webpage
url = driver.get('https://www.naukri.com/hr-recruiters-consultants')

# Finding the search bar and entering the designation to search for
designation = driver.find_element(By.CLASS_NAME, "sugInp")
designation.send_keys('Data Science')

# Clicking the search button
search = driver.find_element(By.XPATH, "/html/body/div[2]/div[2]/div[1]/div[1]/form/div[1]/button")
search.click()

# Creating empty lists to store the scraped data
Name =[]
Designation =[]
Company =[]
Skill =[]
Location =[]

# Scraping Name
try:
    name = driver.find_elements(By.XPATH, '//a[@class="ellipsis"]/span')
    for i in name:
        Name.append(i.text)
except NoSuchElementException:
    Name.append('NA')

# Scraping Designation
try:
    designation = driver.find_elements(By.XPATH, '//p[@class="highlightable"]/span[1]')
    for i in designation:
        Designation.append(i.text)
except NoSuchElementException:
    Designation.append('NA')

# Scraping Company
try:
    company = driver.find_elements(By.XPATH, '//a[@class="ellipsis"][2]')
    for i in company:
        Company.append(i.text)
except NoSuchElementException:
    Company.append('NA')

# Scraping skills they hire for
try:
    skill = driver.find_elements(By.XPATH, '//div[@class="hireSec highlightable"]')
    for i in skill:
        Skill.append(i.text)
except NoSuchElementException:
    Skill.append('NA')

# Scraping Location
try:
    location = driver.find_elements(By.XPATH, '//p[@class="highlightable"]/span[2]/small')
    if location:
        for i in location:
            Location.append(i.text)
    else:
        Location.append('NA')
except NoSuchElementException:
    Location.append('NA')

# Creating a pandas dataframe to store the scraped data
Naukri=pd.DataFrame({})
Naukri['Name']=Name[:49]
Naukri['Designation']=Designation[:49]
Naukri['Company']=Company[:49]
Naukri['Skills I Hire for']=Skill[:49]
Naukri['Location']=Location[:49]
```
